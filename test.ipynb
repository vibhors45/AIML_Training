```python
import pandas as pd
data = pd.read_csv("wine.csv")
y = data['quality']
x = data.drop('quality', axis=1)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```
